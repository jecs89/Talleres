```python
def calculo(lista, porc):
  i = 0
  bananascomidas = 0
  while i < len(lista):
    j = 1
    
    while j <= lista[i]:
      if j % 5 == 0:
        bananascomidas = bananascomidas + 1
      j = j + 1
    i += 1
  return bananascomidas*porc
```
