```python
import requests
```

```python
url = 'https://www.extra.ec/noticia/deportes/ecuador-ciclistas-alistan-recibir-giro-italia-68016.html'
res = requests.get(url)
src = res.text
```

```python
# print(src)
```

```python
identificador = '<h1 class="article-title " id="m31-30-32">'
id = src.find(identificador)
src = src[id+len(identificador):]

identificador = '</h1>'
id = src.find(identificador)
titulo = src[0:id]
print(titulo)
```

```python
def funcion(src):
  identificador = '<h1 class="article-title " id="m31-30-32">'
  id = src.find(identificador)
  src = src[id+len(identificador):]

  identificador = '</h1>'
  id = src.find(identificador)
  titulo = src[0:id]
  # print(titulo)
  
  return titulo
```

```python
titulo1 = funcion(src)
print( titulo1  )
```
